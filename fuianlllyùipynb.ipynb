Now, we will load the model. Please ensure you have sufficient GPU RAM (T4 or higher) available in your Colab runtime.

```python
import torch
from transformers import pipeline
from IPython.display import Audio

# Loading the music generation model
try:
    print("Loading model (facebook/musicgen-small)... this may take a moment.")
    # Using device=0 if a GPU is available for faster generation
    device = 0 if torch.cuda.is_available() else -1
    synthesizer = pipeline("text-to-audio", "facebook/musicgen-small", device=device)

    def generate_music(prompt, duration_seconds=10):
        print(f"Generating {duration_seconds}s of music for: {prompt}")
        # 1 second of audio is roughly 50 tokens
        max_tokens = int(duration_seconds * 50)
        music = synthesizer(prompt, forward_params={"max_new_tokens": max_tokens})
        return music

    # CHANGE YOUR PROMPT HERE
    my_prompt = "upbeat electronic dance music with heavy bass and synthesizers"
    result = generate_music(my_prompt, duration_seconds=10)

    # Display the audio player
    # result['audio'] is a 2D numpy array (channels, samples)
    display(Audio(result["audio"], rate=result["sampling_rate"]))

except Exception as e:
    print(f"An error occurred: {e}")
    print("Tip: Ensure you have selected a GPU runtime (Runtime -> Change runtime type) for better performance.")
```
